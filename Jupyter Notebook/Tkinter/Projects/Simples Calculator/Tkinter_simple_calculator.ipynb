```python
from tkinter import *


def button_click(number):
    entry.insert(END, number)

def button_clear():
    entry.delete(0, END)

def button_back():
    foo = int(len(entry.get()) - 1)
    entry.delete(foo, END)

def button_division():
    global first_num
    global math
    math = "division"
    first_num = int(entry.get())
    entry.delete(0, END)

def button_multiply():
    global first_num
    global math
    math = "multiply"
    first_num = int(entry.get())
    entry.delete(0, END)

def button_subtract():
    global first_num
    global math 
    math = "subtract"
    first_num = int(entry.get())
    entry.delete(0, END)

def button_add():
    global first_num
    global math 
    math = "add"
    first_num = int(entry.get())
    entry.delete(0, END)

def button_equal():
    global math
    second_num = int(entry.get())
    entry.delete(0, END)
    if math == "division":
        entry.insert(0, first_num / second_num)
    elif math == "multiply":
        entry.insert(0, first_num * second_num)
    elif math == "subtract":
        entry.insert(0, first_num - second_num)
    else:
        entry.insert(0, first_num + second_num)
        

root = Tk()
root.title("Simples Calculator")
root.wm_geometry("271x351")
root.wm_resizable('false', 'false')
root.configure(bg = 'red', padx = 2, pady = 2)


entry = Entry(root, borderwidth = 5)
entry.grid(row = 0, column = 0, sticky = W + E, padx = 5, pady = 20, columnspan = 4)

buttom_percent = Button(root, text = "%", padx = 23, pady = 15, command = lambda: button_click(1), state = DISABLED)
buttom_CE = Button(root, text = "CE", padx = 21, pady = 15, command = lambda: button_click(1), state = DISABLED)
buttom_clear = Button(root, text = "C", padx = 24, pady = 15, command = button_clear)
buttom_back = Button(root, text = "<-", padx = 21, pady = 15, command = button_back)

buttom_7 = Button(root, text = 7, padx = 25, pady = 15, command = lambda: button_click(7))
buttom_8 = Button(root, text = 8, padx = 25, pady = 15, command = lambda: button_click(8))
buttom_9 = Button(root, text = 9, padx = 25, pady = 15, command = lambda: button_click(9))
buttom_division = Button(root, text = "/", padx = 25, pady = 15, command = button_division)

buttom_4 = Button(root, text = 4, padx = 25, pady = 15, command = lambda: button_click(4))
buttom_5 = Button(root, text = 5, padx = 25, pady = 15, command = lambda: button_click(5))
buttom_6 = Button(root, text = 6, padx = 25, pady = 15, command = lambda: button_click(6))
buttom_multiply = Button(root, text = "X", padx = 24, pady = 15, command = button_multiply)

buttom_1 = Button(root, text = 1, padx = 25, pady = 15, command = lambda: button_click(1))
buttom_2 = Button(root, text = 2, padx = 25, pady = 15, command = lambda: button_click(2))
buttom_3 = Button(root, text = 3, padx = 25, pady = 15, command = lambda: button_click(3))
buttom_subtract = Button(root, text = "-", padx = 25, pady = 15, command = button_subtract)

buttom_0 = Button(root, text = 0, padx = 25, pady = 15, command = lambda: button_click(0))
buttom_comma = Button(root, text = ", ", padx = 25, pady = 15, command = lambda: button_click(1))
buttom_equal = Button(root, text = "=", padx = 24, pady = 15, command = button_equal)
buttom_add = Button(root, text = "+", padx = 24, pady = 15, command = button_add)


buttom_percent.grid(row = 1, column = 0, padx = 1, pady = 1)
buttom_CE.grid(row = 1, column = 1, padx = 1, pady = 1)
buttom_clear.grid(row = 1, column = 2, padx = 1, pady = 1)
buttom_back.grid(row = 1, column = 3, padx = 1, pady = 1)

buttom_7.grid(row = 2, column = 0, padx = 1, pady = 1)
buttom_8.grid(row = 2, column = 1, padx = 1, pady = 1)
buttom_9.grid(row = 2, column = 2, padx = 1, pady = 1)
buttom_division.grid(row = 2, column = 3, padx = 1, pady = 1)

buttom_4.grid(row = 3, column = 0, padx = 1, pady = 1)
buttom_5.grid(row = 3, column = 1, padx = 1, pady = 1)
buttom_6.grid(row = 3, column = 2, padx = 1, pady = 1)
buttom_multiply.grid(row = 3, column = 3, padx = 1, pady = 1)

buttom_1.grid(row = 4, column = 0, padx = 1, pady = 1)
buttom_2.grid(row = 4, column = 1, padx = 1, pady = 1)
buttom_3.grid(row = 4, column = 2, padx = 1, pady = 1)
buttom_subtract.grid(row = 4, column = 3, padx = 1, pady = 1)

buttom_0.grid(row = 5, column = 0, padx = 1, pady = 1)
buttom_comma.grid(row = 5, column = 1, padx = 1, pady = 1)
buttom_equal.grid(row = 5, column = 2, padx = 1, pady = 1)
buttom_add.grid(row = 5, column = 3, padx = 1, pady = 1)


root.mainloop()
```
