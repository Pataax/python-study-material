```python
from tkinter import *
from PIL import ImageTk, Image


def foward(image_number):
    global show_image
    global current_img
    global button_foward
    global button_back

    if image_number < len(image_list) - 1:
        current_img = image_number
        show_image.grid_forget()
        show_image = Label(image = image_list[image_number])  
        show_image.grid(row = 0, column = 0, columnspan = 3)
        button_foward['state'] = 'normal'
        button_back['state'] = 'normal'
    else:  # quando a image a ser exibida for a ultima. Exibe e desabilita o botao
        current_img = image_number
        show_image.grid_forget()
        show_image = Label(image = image_list[image_number])  
        show_image.grid(row = 0, column = 0, columnspan = 3)
        button_foward['state'] = 'disabled'
        

def back(image_number):
    global show_image
    global current_img
    global button_foward
    global button_back

    if image_number > 0:
        current_img = image_number
        show_image.grid_forget()
        show_image = Label(image = image_list[image_number])  
        show_image.grid(row = 0, column = 0, columnspan = 3)
        button_back['state'] = 'normal'
        button_foward['state'] = 'normal'
    else:
        current_img = image_number
        show_image.grid_forget()
        show_image = Label(image = image_list[image_number])  
        show_image.grid(row = 0, column = 0, columnspan = 3)
        button_back['state'] = 'disabled'



root = Tk()
root.title('Image Viewer')

my_img1 = ImageTk.PhotoImage(Image.open("Images/photo1.png"))
my_img2 = ImageTk.PhotoImage(Image.open("Images/photo2.png"))
my_img3 = ImageTk.PhotoImage(Image.open("Images/photo3.png"))
my_img4 = ImageTk.PhotoImage(Image.open("Images/photo4.png"))
my_img5 = ImageTk.PhotoImage(Image.open("Images/photo5.png"))

image_list = [my_img1, my_img2, my_img3, my_img4, my_img5]

current_img = 0  # inicia o viewer com a primeira imagem aberta
show_image = Label(image = image_list[current_img])  
show_image.grid(row = 0, column = 0, columnspan = 3)

button_back = Button(root, text = "<<", command = lambda: back(current_img - 1))
button_exit = Button(root, text = "Exit Program", command = root.destroy)
button_foward = Button(root, text = ">>", command = lambda: foward(current_img + 1))

button_back.grid(row = 1, column = 0, pady = 10)
button_exit.grid(row = 1, column = 1, pady = 10)
button_foward.grid(row = 1, column = 2, pady = 10)


root.mainloop()
```
